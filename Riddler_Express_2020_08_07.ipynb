```python
#The total distance/length of the trail?
#The middle segment is length 1
#While the four diagonals are hypotenuses of isosceles rights with side length 1
#Therefore there are four distances with length root(2)
#So precisely, in miles, the length of the trail is 4*pos.root(2) + 1
```

```python
math.sqrt(2)*4 + 1
```

```python
#Extra Credit: Al-Battani Island
#Some necessary math for the ellipse: 
#See the bottom of this page: 
#https://www.mathsisfun.com/geometry/ellipse.html

#We center the ellipse at 0, 0 for convenient algebra

#We want the coordinates (x, y) on the ellipse where the points on the major axis
#directly above/below that point
#are equally distant from the shore 
#by travelling along the major axis
#or by going to the point

#The trivial solution is the terminus of the major axis; 
#we want the other one

#To solve the quadratic equation in Python: 
#Source: 
#https://www.programiz.com/python-programming/examples/quadratic-roots

a = 3.25/2.25
b = -3
c = 1.25

# calculate the discriminant
d = (b**2) - (4*a*c)

# find two solutions
sol1 = (-b-cmath.sqrt(d))/(2*a)
sol2 = (-b+cmath.sqrt(d))/(2*a)

print('The solutions are {0} and {1}'.format(sol1,sol2))
```

```python
#Now for the elliptical island: 

fig = plt.figure()
ax = fig.add_subplot(111)

#Draw the perfectly elliptical island and add it:  
rect1 = matplotlib.patches.Ellipse(xy = (0, 0), width = 3, height = 2, color='#dec678')
ax.add_patch(rect1)

#And we set a Mediterranean-blue background: 
#https://stackoverflow.com/questions/14088687/how-to-change-plot-background-color
ax.set_facecolor('#5695d1')

#Adding some lines: 
#https://stackoverflow.com/questions/36470343/how-to-draw-a-line-with-matplotlib/36479941

#Here are the major and minor axes: 

#The x coords for the first line
x1 = [-1.5, 1.5]
#The y coords for the first line
y1 = [0, 0]
#I know. That doesn't make any sense either.

#Coords for the second line, x ones then y ones
x2 = [0, 0]
y2 = [-1, 1]

#We want this value after working out the equation of an ellipse: 
#See above
keyvariable = sol1.real

ycomp = 1.5 - keyvariable

x3 = [-keyvariable, keyvariable]
y3 = [0, 0]

x4 = [-keyvariable, -keyvariable]
y4 = [ycomp, -ycomp]

x5 = [keyvariable, keyvariable]
y5 = [ycomp, -ycomp]

#Drawing the lines
#cf https://www.kite.com/python/answers/how-to-plot-different-colored-lines-in-matplotlib-in-python
#axes in pink
plt.plot(x1, y1, marker = '.', color = 'pink')
plt.plot(x2, y2, marker = '.', color = 'pink')
#trail in red
plt.plot(x3, y3, marker = '.', color = 'red')
#dropped altitudes in green
plt.plot(x4, y4, marker = '.', color = '#4cc26b')
plt.plot(x5, y5, marker = '.', color = '#4cc26b')

plt.xlim([-2.5, 2.5])
plt.ylim([-2, 2])
plt.show()
```

```python
#And therefore the length of the trail is: 
keyvariable * 2
```
