```python
#IMPORTS
#We will need pi for the formula for the surface area of a sphere
#https://en.wikipedia.org/wiki/Sphere#Surface_area
#So we import the variable in the math package: 
import math
#math.pi
```

We use the formula for the surface area of a sphere (given its radius):

<center>$A = 4\pi r^{2}$</center>

```python
#For modularity, we define a function to return a sphere's surface area: 
def sphere_area(rad=1):
    return 4*math.pi*(rad**2)
```

```python
#This problem defines the earth as a perfect sphere with radius thus, in KM: 
earth_rad = 6378
#And asks for the difference in surface area between this earth
#And a sphere 1 meter (1/1000 KM) off the ground from it: 
#So the answer in KM is: 
print(sphere_area(earth_rad+0.001) - sphere_area(earth_rad))
```

There's [a lot going on](https://en.wikipedia.org/wiki/List_of_lunar_features#:~:text=The%20surface%20of%20the%20Moon,of%20these%20features%20are%20listed.) on the moon. And though we lack exact measurements for many of the maria, laci, and paludes that decorate her surface, we can find a range of measurements within which there might be an ever more precise equivalent to represent the extra surface of sheet we'd need for our planet . . . 

The Wikipedia page lists sizes in terms of diameters, rounded to the nearest kilometer, so here's a function to take those values and give us our range of promise: 

```python
def moon_area(diameter = 1):
    return math.pi*(diameter/2)**2

print(moon_area(14))
print(moon_area(15))
```
